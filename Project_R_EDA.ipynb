# **International Year of Millets 2023**

## United Nations General Assembly declared 2023 as the International Year of Millets.

The International Year of Millet gives a thriving opportunity to 
1. Increase the contribution of millet to food security.
2. Increase the global production of millets
3. Ensuring efficient processing,transport,storage, and consumption.
4. Sustainable production and quality of millet with the involvement of the stakeholder.

Indian Millet Sourcing Point :
India is one of the leading producers and suppliers of millet .
The main millet-growing states in India are Rajasthan,Maharashtra,Karnataka,AndhraPradesh,MadhyaPradesh.

Millets are small-grained, Jowar(Sorghum),Bajra(Pearl Millet) and Ragi(Finger Millet) are the important millets cultivated in India.

```r
library("ggplot2")
```

```r
ggplot(df, aes(x = Crop, y = Yield..Kg.per.Ha..)) +
  geom_boxplot(fill = "orange") +
  labs(title = "Yield by Crop", x = "Crop", y = "Yield (Kg per Ha)")
```

```r
ggplot(df, aes(x =Yield..Kg.per.Ha.., y = Production....000.Tonnes., color = State)) +
  geom_point() +
  labs(title = "Production vs. Yield by State", x = "Yield (Kg per Ha)", y = "Production (Tonnes)")
```

```r
df_summary <- df %>%
  group_by(Crop) %>%
  summarise(Avg_Yield = mean(Yield..Kg.per.Ha..))

# bar chart of Average Yield by Crop
ggplot(df_summary, aes(x = Crop, y = Avg_Yield)) +
  geom_bar(stat = "identity", fill = "darkblue") +
  labs(title = "Average Yield by Crop", x = "Crop", y = "Average Yield (Kg per Ha)")
```
